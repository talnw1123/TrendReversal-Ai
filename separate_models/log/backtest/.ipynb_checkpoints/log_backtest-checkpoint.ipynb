# Backtest Results — Best Specialized Models (Enhanced MOO Selection)
**Script:** `model/backtest_all_models.py`  
**Source:** Actual run log (`log_backtest.txt`)  
**Markets:** BTC, Gold, Thai, UK, US (× Uptrend / Downtrend = 10 backtests)  
**Optimization:** Discrete modes + Weighted (confidence/asymmetric/pyramid) + EMA smoothing + Vol targeting + Trailing stops  
**Scoring:** Alpha + 0.5×Sortino + 2.0×Calmar − DD penalty

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['figure.dpi'] = 130
print('Libraries loaded')
```

## 1. Load Backtest Results from Actual Run Log

```python
# From log_backtest.txt FINAL RESULTS table (latest run with enhanced strategies)
raw = [
    {'Market':'US',  'Trend':'uptrend',  'Regime':'HMM',                'Model':'RandomForest','Mode':'standard',                       'Threshold':0.50,'Return':34.01,'BnH':33.11,'MaxDD':-18.90,'Final':13400.82,'Trades':471,'Skipped':0},
    {'Market':'US',  'Trend':'downtrend','Regime':'HMM',                'Model':'CNN',         'Mode':'W(asymmetric,ema3,f0.3,vt0.00)', 'Threshold':0.55,'Return':33.28,'BnH':33.11,'MaxDD':-13.74,'Final':13327.83,'Trades':471,'Skipped':0},
    {'Market':'UK',  'Trend':'uptrend',  'Regime':'HMM',                'Model':'CNN',         'Mode':'long_bias',                      'Threshold':0.50,'Return':27.09,'BnH':27.21,'MaxDD':-13.43,'Final':12708.56,'Trades':477,'Skipped':0},
    {'Market':'UK',  'Trend':'downtrend','Regime':'HMM',                'Model':'XGBoost',     'Mode':'long_bias',                      'Threshold':0.80,'Return':27.09,'BnH':27.21,'MaxDD':-13.43,'Final':12708.56,'Trades':0,  'Skipped':477},
    {'Market':'Thai','Trend':'uptrend',  'Regime':'LogisticRegression', 'Model':'LSTM',        'Mode':'long_only',                      'Threshold':0.64,'Return':13.98,'BnH':15.13,'MaxDD':-7.48, 'Final':11397.82,'Trades':149,'Skipped':303},
    {'Market':'Thai','Trend':'downtrend','Regime':'LogisticRegression', 'Model':'CNN',         'Mode':'bh_anchor',                      'Threshold':0.50,'Return':15.01,'BnH':15.13,'MaxDD':-28.91,'Final':11501.35,'Trades':452,'Skipped':0},
    {'Market':'Gold','Trend':'uptrend',  'Regime':'SMA200',             'Model':'CNN',         'Mode':'long_bias',                      'Threshold':0.50,'Return':100.65,'BnH':100.85,'MaxDD':-17.73,'Final':20065.11,'Trades':472,'Skipped':0},
    {'Market':'Gold','Trend':'downtrend','Regime':'SMA200',             'Model':'Transformer', 'Mode':'long_bias',                      'Threshold':0.74,'Return':100.65,'BnH':100.85,'MaxDD':-17.73,'Final':20065.11,'Trades':0,  'Skipped':472},
    {'Market':'BTC', 'Trend':'uptrend',  'Regime':'XGBoost',            'Model':'RandomForest','Mode':'long_only',                      'Threshold':0.60,'Return':21.23,'BnH':15.63,'MaxDD':-18.17,'Final':12123.32,'Trades':90, 'Skipped':611},
    {'Market':'BTC', 'Trend':'downtrend','Regime':'XGBoost',            'Model':'RandomForest','Mode':'W(pyramid,ema10,f0.2,vt0.00)',   'Threshold':0.55,'Return':63.41,'BnH':15.63,'MaxDD':-35.31,'Final':16341.45,'Trades':400,'Skipped':301},
]

df = pd.DataFrame(raw)
df['Alpha'] = df['Return'] - df['BnH']
df['Beat_BnH'] = df['Alpha'] > 0
df['Label'] = df['Market'] + '\n(' + df['Trend'].str[:4] + ')'

# Simplified mode for plotting (long names break charts)
def simplify_mode(m):
    if m.startswith('W('):
        # extract sizing type
        if 'pyramid' in m: return 'pyramid'
        if 'asymmetric' in m: return 'asymmetric'
        if 'confidence' in m: return 'confidence'
        return 'weighted'
    return m
df['ModeShort'] = df['Mode'].apply(simplify_mode)

print(f'Total backtests: {len(df)}')
print(f'Beat B&H: {df["Beat_BnH"].sum()} / {len(df)} ({df["Beat_BnH"].mean()*100:.1f}%)')
print(f'Average Alpha: {df["Alpha"].mean():+.2f}%')
df[['Market','Trend','Model','Mode','Threshold','Return','BnH','Alpha','MaxDD']]
```

## 2. Strategy Return vs Buy & Hold

```python
fig, ax = plt.subplots(figsize=(14, 6))
x = np.arange(len(df))
w = 0.38

b1 = ax.bar(x - w/2, df['Return'], w, label='Strategy Return', color='#2E86AB', edgecolor='white')
b2 = ax.bar(x + w/2, df['BnH'],    w, label='Buy & Hold',     color='#E07A5F', edgecolor='white')

for bar, val in zip(b1, df['Return']):
    ax.text(bar.get_x() + bar.get_width()/2, val + (1.5 if val>=0 else -3),
            f'{val:+.1f}%', ha='center', fontsize=8.5, fontweight='bold', color='#1B4965')
for bar, val in zip(b2, df['BnH']):
    ax.text(bar.get_x() + bar.get_width()/2, val + 1.5,
            f'{val:+.1f}%', ha='center', fontsize=8.5, color='#7A2E1F')

ax.set_xticks(x)
ax.set_xticklabels(df['Label'], fontsize=10)
ax.axhline(0, color='black', linewidth=0.7)
ax.set_ylabel('Return (%)', fontsize=12)
ax.set_title('Strategy Return vs Buy & Hold — Per Market × Trend', fontsize=14, fontweight='bold')
ax.legend(fontsize=11, loc='upper left')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('return_vs_bnh.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Alpha — Excess Return Over Buy & Hold

```python
fig, ax = plt.subplots(figsize=(13, 5))
colors = ['#2E7D32' if a > 0 else '#C62828' for a in df['Alpha']]
bars = ax.bar(df['Label'], df['Alpha'], color=colors, edgecolor='white', linewidth=1.2)

for bar, a in zip(bars, df['Alpha']):
    offset = 1.0 if a >= 0 else -2.0
    ax.text(bar.get_x() + bar.get_width()/2, a + offset,
            f'{a:+.2f}%', ha='center', fontsize=10, fontweight='bold')

ax.axhline(0, color='black', linewidth=0.8)
ax.set_ylabel('Alpha = Strategy − B&H (%)', fontsize=12)
ax.set_title('Alpha — Excess Return vs Buy & Hold\n(Green = Beat B&H, Red = Underperform)', fontsize=13, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('alpha_vs_bnh.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\nBeat B&H: {df["Beat_BnH"].sum()} / {len(df)} ({df["Beat_BnH"].mean()*100:.0f}%)')
print(f'Average Alpha: {df["Alpha"].mean():+.2f}%')
print(f'Best Alpha: {df["Alpha"].max():+.2f}% ({df.loc[df["Alpha"].idxmax(), "Market"]} {df.loc[df["Alpha"].idxmax(), "Trend"]})')
print(f'Worst Alpha: {df["Alpha"].min():+.2f}% ({df.loc[df["Alpha"].idxmin(), "Market"]} {df.loc[df["Alpha"].idxmin(), "Trend"]})')
```

## 4. Risk Profile — Maximum Drawdown

```python
fig, ax = plt.subplots(figsize=(13, 5))
dd_colors = plt.cm.RdYlGn_r((df['MaxDD'].abs() / df['MaxDD'].abs().max()).values)
bars = ax.bar(df['Label'], df['MaxDD'], color=dd_colors, edgecolor='white', linewidth=1.2)

for bar, dd in zip(bars, df['MaxDD']):
    ax.text(bar.get_x() + bar.get_width()/2, dd - 1.0,
            f'{dd:.2f}%', ha='center', va='top', fontsize=10, fontweight='bold', color='black')

ax.set_ylabel('Maximum Drawdown (%)', fontsize=12)
ax.set_title('Maximum Drawdown per Market × Trend (Lower magnitude = Safer)', fontsize=13, fontweight='bold')
ax.set_ylim(df['MaxDD'].min() * 1.15, 5)
ax.axhline(0, color='black', linewidth=0.8)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('max_drawdown.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Risk vs Return Scatter

```python
fig, ax = plt.subplots(figsize=(10, 7))

market_colors = {'BTC':'#F2A900','Gold':'#FFD700','Thai':'#0088CC','UK':'#012169','US':'#B22234'}
trend_marker = {'uptrend':'^','downtrend':'v'}

for _, r in df.iterrows():
    ax.scatter(abs(r['MaxDD']), r['Return'],
               s=300, c=market_colors[r['Market']],
               marker=trend_marker[r['Trend']],
               edgecolors='black', linewidths=1.5, alpha=0.85, zorder=3)
    ax.annotate(f"{r['Market']} {r['Trend'][:2].upper()}\n{r['ModeShort']}",
                (abs(r['MaxDD']), r['Return']),
                xytext=(8, 5), textcoords='offset points', fontsize=8.5)

ax.axhline(0, color='gray', linestyle='--', linewidth=1, alpha=0.6)
ax.set_xlabel('|Max Drawdown| (%)  — Risk →', fontsize=12)
ax.set_ylabel('Strategy Return (%)  — Reward →', fontsize=12)
ax.set_title('Risk vs Return — Best Strategy per Market\n▲ Uptrend  ▼ Downtrend', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)

patches = [mpatches.Patch(color=c, label=m) for m, c in market_colors.items()]
ax.legend(handles=patches, loc='upper right', title='Market')
plt.tight_layout()
plt.savefig('risk_return_scatter.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Selected Strategy Mode — Distribution

```python
mode_count = df['ModeShort'].value_counts().reset_index()
mode_count.columns = ['Mode', 'Count']

mode_palette = {
    'long_bias':'#2E86AB','long_only':'#A23B72','standard':'#F18F01',
    'bh_anchor':'#C73E1D','asymmetric':'#3B6E8F','pyramid':'#7B2D26',
    'confidence':'#5D737E','long_neutral':'#888888','weighted':'#888888'
}

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.barh(mode_count['Mode'], mode_count['Count'],
               color=[mode_palette.get(m, '#888') for m in mode_count['Mode']],
               edgecolor='white', linewidth=1)

for bar, cnt in zip(bars, mode_count['Count']):
    ax.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
            f'{cnt} case(s)', va='center', fontsize=11, fontweight='bold')

ax.set_xlabel('Number of Cases Selected', fontsize=11)
ax.set_title('Strategy Mode Selection Distribution\n(Across 10 Market × Trend combinations)',
             fontsize=13, fontweight='bold')
ax.set_xlim(0, mode_count['Count'].max() + 1.5)
ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('mode_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n=== Mode Selection Summary ===')
print(mode_count.to_string(index=False))
```

## 7. Trade Activity vs Threshold

```python
fig, ax = plt.subplots(figsize=(13, 5.5))
x = np.arange(len(df))
ax.bar(x, df['Trades'],  label='Trades Executed', color='#388E3C', edgecolor='white')
ax.bar(x, df['Skipped'], bottom=df['Trades'],
       label='Skipped (Low Confidence)', color='#9E9E9E', alpha=0.7, edgecolor='white')

for i, r in df.iterrows():
    total = r['Trades'] + r['Skipped']
    if total > 0:
        pct = r['Trades'] / total * 100
        ax.text(i, total + 8, f"{pct:.0f}% active\n(Th={r['Threshold']:.2f})",
                ha='center', fontsize=8.5, fontweight='bold')

ax.set_xticks(x)
ax.set_xticklabels(df['Label'], fontsize=10)
ax.set_ylabel('Number of Signals', fontsize=12)
ax.set_title('Trade Activity — Optimized Threshold per Market', fontsize=13, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('trade_activity.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 8. Final Capital from $10,000 Initial

```python
fig, ax = plt.subplots(figsize=(13, 5.5))
df_sorted = df.sort_values('Final', ascending=True).reset_index(drop=True)
colors = ['#388E3C' if v >= 10000 else '#C62828' for v in df_sorted['Final']]
bars = ax.barh(df_sorted['Label'].str.replace('\n',' '), df_sorted['Final'],
               color=colors, edgecolor='white', linewidth=1.2)

for bar, val, model, mode in zip(bars, df_sorted['Final'], df_sorted['Model'], df_sorted['ModeShort']):
    ax.text(bar.get_width() + 200, bar.get_y() + bar.get_height()/2,
            f'${val:,.0f}  ({model}, {mode})', va='center', fontsize=9.5, fontweight='bold')

ax.axvline(10000, color='black', linestyle='--', linewidth=1.5, label='Initial $10,000')
ax.set_xlabel('Final Capital (USD)', fontsize=12)
ax.set_title('Final Portfolio Value — Initial $10,000\n(Green = Profit, Red = Loss)',
             fontsize=13, fontweight='bold')
ax.legend(fontsize=11, loc='lower right')
ax.grid(axis='x', alpha=0.3)
ax.set_xlim(0, df_sorted['Final'].max() * 1.20)
plt.tight_layout()
plt.savefig('final_capital.png', dpi=150, bbox_inches='tight')
plt.show()
```
